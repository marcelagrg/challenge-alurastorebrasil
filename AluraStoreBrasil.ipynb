### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
faturamento1 = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()
```

```python
print('Faturamento Loja 1:', faturamento1)
print('Faturamento Loja 2:',faturamento2)
print('Faturamento Loja 3:',faturamento3)
print('Faturamento Loja 4:',faturamento4)
```

```python
import matplotlib.pyplot as plt

x = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
y = [faturamento1, faturamento2, faturamento3, faturamento4]

plt.bar(x, y, color='greenyellow')
plt.ylim(1000000, 1600000)
plt.ylabel('Faturamento em milhões de reais')

plt.show()
```

# 2. Vendas por Categoria

```python
venda_por_categoria_loja1 = loja['Categoria do Produto'].value_counts()

print('Vendas por categoria Loja 1:')
display(venda_por_categoria_loja1)
```

```python
categorias = loja['Categoria do Produto'].value_counts().index
quantidade = loja['Categoria do Produto'].value_counts().values

plt.barh(categorias, quantidade, color='orange')
plt.xlabel('Quantidade')
plt.ylabel('Categoria do Produto')
plt.title('Vendas por categoria Loja 1')

plt.show()
```

```python
venda_por_categoria_loja2 = loja2['Categoria do Produto'].value_counts()

print('Vendas por categoria Loja 2 :')
display(venda_por_categoria_loja2)
```

```python
venda_por_categoria_loja3 = loja3['Categoria do Produto'].value_counts()

print('Vendas por categoria Loja 3:')
display(venda_por_categoria_loja3)
```

```python
venda_por_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print('Vendas por categoria Loja 4:')
display(venda_por_categoria_loja4)
```

```python
soma_total_categorias = venda_por_categoria_loja1 + venda_por_categoria_loja2 + venda_por_categoria_loja3 + venda_por_categoria_loja4

labels = soma_total_categorias.index
sizes = soma_total_categorias.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Total de vendas por categoria')

plt.show()
```

# 3. Média de Avaliação das Lojas


```python
media_loja1 = loja['Avaliação da compra'].mean().round(2)
media_loja2 = loja2['Avaliação da compra'].mean().round(2)
media_loja3 = loja3['Avaliação da compra'].mean().round(2)
media_loja4 = loja4['Avaliação da compra'].mean().round(2)
```

```python
print('Média de avaliação Loja 1:', media_loja1)
print('Média de avaliação Loja 2:', media_loja2)
print('Média de avaliação Loja 3:', media_loja3)
print('Média de avaliação Loja 4:', media_loja4)
```

```python
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento = [faturamento1, faturamento2, faturamento3, faturamento4]
avaliacao = [media_loja1, media_loja2, media_loja3, media_loja4]

fig, ax1 = plt.subplots()

ax1.bar(lojas, faturamento, color='skyblue', label='Faturamento (Milhão R$)')
ax1.set_xlabel('Lojas')
ax1.set_ylabel('Faturamento (Milhão R$)', color='black')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(1000000, 1600000)

ax2 = ax1.twinx()
ax2.plot(lojas, avaliacao, color='red', marker='o', linewidth=2, label='Avaliação Média')
ax2.set_ylabel('Avaliação Média', color='black')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylim(3.5, 5)

plt.title('Faturamento x Avaliação Média por Loja')

fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```

##Introdução
Este relatório apresenta uma análise do desempenho das quatro lojas **Alura Store**, pertencentes ao Senhor João, com o objetivo de identificar aquela com **menor eficiência** e, consequentemente, mais indicada para ser vendida.

O que foi analisado:
- Faturamento total
- Média das avaliações de clientes
- Frete médio
- Categorias de produtos
- Produtos mais e menos vendidos


---


##Faturamento Total
- A Loja 1 apresenta o maior faturamento, com vantagem significativa.
- A Loja 4 possui o menor faturamento, ficando R\$ 79.527,45 abaixo da Loja 3 (a segunda com pior desempenho) e R\$ 150.011,54 abaixo da Loja 1.
---


##Avaliação Média
- Baixa diferença entre a média de avaliações, indicando um padrão de satisfação dos clientes semelhante em todas as lojas.
- A Loja 3 possui a melhor avaliação, com nota média de 4,05.
- A Loja 1 registra a menor avaliação média, com 3,98.
---

##Frete Médio
- O frete não exerce grande influência sobre os desempenhos das lojas.
- A Loja 1, mesmo com o maior frete médio, possui o maior faturamento.
- A Loja 4 apresenta o menor frete médio, mas ainda assim tem o pior faturamento.
---

##Vendas por categoria
- As quatro lojas apresentam exatamente o mesmo Top 3 de categorias mais vendidas, indicando um padrão de consumo consistente entre os clientes:

  1. Móveis

  2. Eletrônicos

  3. Brinquedos


---

##Vendas por produtos
- A Loja 1 apresenta alta demanda por produtos grandes das categorias de móveis e eletrodomésticos, enquanto brinquedos apresentam baixa procura.

- Nenhum produto das três categorias mais vendidas aparece no Top 5 da Loja 2. Destaca-se o fato de seu item mais vendido ser um livro, acompanhado por três instrumentos musicais, enquanto eletrônicos tem menor demanda. Esse cenário sugere que a loja tem forte saída de itens específicos, enquanto as categorias mais populares não concentram suas vendas em poucos produtos.

- A Loja 3 apresenta grande vendas de móveis, mostrando consistência no perfil dos compradores. Os itens menos vendidos são de categorias variadas, sem padrão definido.

- Destaca-se a presença de dois tipos de camas nos produtos mais vendidos da Loja 4, porém, o restante do Top 5 apresenta itens de categorias variadas. Os produtos com menor demanda também são diversos, com destaque para instrumentos musicais. A presença de um livro entre os mais vendidos e outro entre os menos vendidos demonstra baixa consistência no padrão de demanda por categorias.

---

##Conclusão
Com base nos indicadores avaliados, a loja com menor eficiência e opção mais adequada para ser vendida é a **Loja 4**. Os principais motivos são:
1. Loja com menor faturamento, indicando baixo número de vendas ou valor médio por produtos.
2. Segunda pior avaliação média, ficando atrás apenas da Loja 1, que apresenta um faturamento R\$ 150.011,54 maior que a Loja 4.
3. Apresenta o menor valor de frete médio, porém isso não exerce influência significativa nas vendas.
4. Padrão de vendas pouco consistente por categoria, diferentemente das demais lojas, que apresentam perfis mais claros e previsíveis quanto aos itens mais vendidos.
